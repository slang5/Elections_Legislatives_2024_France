```python
import scipy as sc
import numpy as np
import pandas as pd
import matplotlib.pyplot as mp
import seaborn as sb
```

```python
DataBase = pd.read_excel('DataForModelTrain.xlsx')
DataBase
```

```python
Ytrain = DataBase[["Gagnant Legi T2 2022"]]
```

```python
Ytrain
```

```python
Xtrain = DataBase.drop(["NumDep_Circi","% Voix/Exp Gagnant","Groupe gagnant","Gagnant Legi T2 2022"],axis=1)
```

```python
Xtrain
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
Xtrain_norm = scaler.fit_transform(Xtrain)
```

```python
from sklearn.kernel_ridge import KernelRidge
knlridge = KernelRidge(alpha=1,gamma=1.1,degree=500)
```

```python
knlridge.fit(Xtrain,Ytrain)
```

```python
knlridge.score(X=Xtrain,y=Ytrain)
```

```python
Ypredict = knlridge.predict(Xtrain)
rounded_predictions = np.round(Ypredict).astype(int)
ypred =[1]*568
```

```python
count =0
for i in rounded_predictions:
    if i <=1:
        ypred[count] = 1
    elif i>=4:
        ypred[count] =4
    else:
        ypred[count] = i[0]
    count +=1
```

```python
from sklearn.metrics import recall_score

recall_score(Ytrain,ypred,average='micro')
```

Le modèle Kernel Ridge est à 63,2%

```python
from sklearn.cluster import KMeans

AlgoKmeans = KMeans(n_clusters=4,max_iter=3000,random_state=12,tol=0.001)
AlgoKmeans.fit(Xtrain_norm)
AlgoKmeans.score(Xtrain_norm,Ytrain)
```

```python
ypredictKmeans = AlgoKmeans.predict(Xtrain_norm)
for i in range(0,len(ypredictKmeans)):
    ypredictKmeans[i]=ypredictKmeans[i]+1
ypredictKmeans
```

```python
recall_score(Ytrain,ypredictKmeans,average="micro")
```

en k-mean le score est de 23,7%

```python
from sklearn.cluster import SpectralClustering

Model_SpectralClustering = SpectralClustering(random_state=12,n_clusters=70,n_components=4,n_init=20,assign_labels='cluster_qr')
Model_SpectralClustering.fit(Xtrain_norm)
```

```python
ypredictModel_SpectralClustering = Model_SpectralClustering.fit_predict(Xtrain_norm)
```

```python
for i in range(0,len(ypredictModel_SpectralClustering)):
    ypredictModel_SpectralClustering[i]=ypredictModel_SpectralClustering[i]+1
```

```python
ypredictModel_SpectralClustering
```

```python
recall_score(Ytrain,ypredictModel_SpectralClustering,average="micro")
```

le modèle SpectralClustering fait un score de 22%

```python
from sklearn.tree import ExtraTreeClassifier

Model_ExtraTreeClassifier = ExtraTreeClassifier(splitter="best",random_state=12,max_depth=5,max_features=None)
Model_ExtraTreeClassifier.fit(Xtrain_norm,Ytrain)
Model_ExtraTreeClassifier.score(Xtrain_norm,Ytrain)
```

```python
featimprot = Model_ExtraTreeClassifier.feature_importances_
column = Xtrain.columns
```

```python
DataBaserandom = DataBase.sample(frac=1,axis=0)
DataBaserandom = DataBaserandom.reset_index()
DataBaserandom
```

```python
Y = DataBaserandom["Gagnant Legi T2 2022"]
X = DataBaserandom.drop(["index","NumDep_Circi","% Voix/Exp Gagnant","Groupe gagnant","Gagnant Legi T2 2022"],axis=1)
```

```python
N = 350
XrandomTrain,XrandomTest = X.iloc[:N],X.iloc[N:]
YrandomTrain,YrandomTest = Y.iloc[:N],Y.iloc[N:]
```

```python
Treeclassifier = ExtraTreeClassifier(random_state=12,max_depth=4,max_features=None,splitter="best")
Treeclassifier.fit(XrandomTrain,YrandomTrain)
scoreTrain = Treeclassifier.score(XrandomTrain,YrandomTrain)
scoreTest = Treeclassifier.score(XrandomTest,YrandomTest)
print(f"Train : {scoreTrain} ; Test : {scoreTest}")
```

```python
LogisRegr = LogisticRegression(random_state=12,max_iter=10000,solver='liblinear')
LogisRegr.fit(np.array(XrandomTrain),np.array(YrandomTrain))
scoreTrainLogis = LogisRegr.score(XrandomTrain,YrandomTrain)
scoreTestLogis = LogisRegr.score(XrandomTest,YrandomTest)
print(f"Train : {scoreTrainLogis} ; Test : {scoreTestLogis}")
```

```python
from sklearn.linear_model import SGDClassifier
Model_SGDClassifier = SGDClassifier(max_iter=100000,random_state=12,loss="log_loss")
Model_SGDClassifier.fit(XrandomTrain,YrandomTrain)
scoreTrainSGDClassifier = Model_SGDClassifier.score(XrandomTrain,YrandomTrain)
scoreTestSGDClassifier = Model_SGDClassifier.score(XrandomTest,YrandomTest)
print(f"Train : {scoreTrainSGDClassifier} ; Test : {scoreTestSGDClassifier}")
```

```python
from sklearn.svm import LinearSVC
Model_LinearSVC = LinearSVC(max_iter=99999999,random_state=12,dual=False)
Model_LinearSVC.fit(XrandomTrain,YrandomTrain)
scoreTrainModel_LinearSVC = Model_LinearSVC.score(XrandomTrain,YrandomTrain)
scoreTestModel_LinearSVC = Model_LinearSVC.score(XrandomTest,YrandomTest)
print(f"Train : {scoreTrainModel_LinearSVC} ; Test : {scoreTestModel_LinearSVC}")
```

```python
from sklearn.svm import SVC
model_SVC = SVC(random_state=12,max_iter=999999999,degree=5,kernel='poly')
model_SVC.fit(XrandomTrain,YrandomTrain)
scoreTrainmodel_SVC = model_SVC.score(np.array(XrandomTrain),np.array(YrandomTrain))
scoreTestmodel_SVC = model_SVC.score(XrandomTest,YrandomTest)
print(f"Train : {scoreTrainmodel_SVC} ; Test : {scoreTestmodel_SVC}")
```
